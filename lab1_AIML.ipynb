```python
#part2
people=['klaus','damon','elena','stefan','bonnie']
age=[35,36,37,40,39]
weight=[55,60,50,62,55]
height=[160,155,140,162,150]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("Relationship btn wt and ht of patients")
plt.ylabel("ht in meters")
plt.xlabel("wt in kgs")
plt.show()
```

```python
plt.scatter(age,height)
plt.title("Relationship btn age and ht of patients")
plt.ylabel("ht in meters")
plt.xlabel("age")
plt.show()
```

```python
plt.scatter(weight,age)
plt.title("Relationship btn wt and age of patients")
plt.ylabel("age")
plt.xlabel("wt in kgs")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
#main plot function bar
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
#plt.ticks(fontsize=14,rotation=30)
#plt.yticks(fomtsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
#main plot function"hist"
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
#plt.ticks(fontsize=15)
#plt.yticks(fontsize=15)
plt.show()
```
